#Visualize Time data C101.txt

```python
import matplotlib.pyplot as plt

# 1. Trích xuất dữ liệu từ bảng kết quả benchmark (C101.txt)
customers = [5, 6, 7, 8, 9, 10, 11]

time_bb = [0.0458, 0.4929, 0.4638, 1.0280, 7.9717, 238.4729, 473.9168]
time_bc = [0.0324, 0.0436, 0.0908, 0.1325, 0.1347, 16.6420, 77.7982]
time_bp = [0.0136, 0.0237, 0.0300, 0.0683, 0.1471, 0.3317, 0.5037]

# 2. Khởi tạo biểu đồ
plt.figure(figsize=(10, 6))

# 3. Vẽ các đường cho từng phương pháp
plt.plot(customers, time_bb, marker='o', linestyle='-', color='red', linewidth=2, label='Branch & Bound (B&B)')
plt.plot(customers, time_bc, marker='s', linestyle='-', color='orange', linewidth=2, label='Branch & Cut (B&C)')
plt.plot(customers, time_bp, marker='^', linestyle='-', color='green', linewidth=2, label='Branch & Price (B&P)')

# 4. Định dạng biểu đồ
plt.title('Comparison of Execution Time by Number of Customers (Dataset: C101)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Number of Customers', fontsize=12)
plt.ylabel('Execution Time (Seconds) - Log Scale', fontsize=12)

# Sử dụng thang đo Logarit cho trục Y
plt.yscale('log')

# Thiết lập các mốc trục X là số nguyên
plt.xticks(customers)

# Tắt hoàn toàn các mốc chia phụ (minor ticks) trên các trục
plt.minorticks_off()

# Chỉ hiện lưới ở các mốc chính (major), và chỉ kẻ vạch ngang (axis="y")
plt.grid(True, which="major", axis="y", ls="--", alpha=0.5)

# Hiển thị chú thích (legend)
plt.legend(fontsize=11, loc='upper left')

# 5. Căn chỉnh và hiển thị/lưu biểu đồ
plt.tight_layout()

# Nếu muốn lưu ảnh ra file, hãy bỏ comment dòng bên dưới:
# plt.savefig('time_comparison_clean.png', dpi=300)

plt.show()
```
